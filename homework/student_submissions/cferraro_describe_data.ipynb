```python
import numpy as np
import scipy.stats as stats
import csv
import seaborn as sns

%matplotlib inline
```

```python
rows = []
```

```python
sales_csv_path = './sales_info.csv'
```

```python
with open(sales_csv_path, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        rows.append(row)
f.close()
```
