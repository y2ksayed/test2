```python
#  Neuaral Network
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from keras.models import Sequential
from keras.layers import Dense

feature_cols = ['YRS_TO_MTY_ISSUE','BB_5Y_DEFAULT_PROB','YRS_TO_MTY_ISSUE','DUR_MID','rating_num_A','rating_num_A+','rating_num_A-','rating_num_AA','rating_num_AA+','rating_num_AA+','rating_num_AA+','rating_num_AA+','rating_num_AA+','rating_num_AA+',
                'rating_num_AA-','rating_num_AAA','rating_num_AAA','rating_num_BB','rating_num_BB+','rating_num_BBB',
                'rating_num_BBB','rating_num_BBB+','rating_num_BBB-','rating_num_BBB-','industry_num_Basic Materials','industry_num_Communications',
                'industry_num_Consumer, Cyclical','industry_num_Consumer, Non-cyclical','industry_num_Energy','industry_num_Financial','industry_num_Government',
                'industry_num_Industrial','industry_num_Technology','industry_num_Utilities','AMT_OUTSTANDING',
                'LQA_LIQUIDITY_SCORE','LQA_BID_ASK_SPREAD','BB_1YR_DEFAULT_PROB','BB_5Y_DEFAULT_PROB']

X = list[feature_cols]
y = list.YAS_OAS_SPRD

X_train, X_test, y_train, y_test = train_test_split(X, y)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)


n_input = X_train.shape[1]
n_input
```

```python

model = Sequential()
model.add(Dense(39, input_dim=n_input, activation='relu'))
model.add(Dense(6, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, activation='relu'))
model.compile(loss='mse', optimizer='rmsprop')

history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=500,
    batch_size=None,
    verbose=2
)
```

```python
from keras.callbacks import EarlyStopping


model = Sequential()
model.add(Dense(39, input_dim=n_input, activation='relu'))
model.add(Dense(6, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, activation='relu'))
model.compile(loss='mse', optimizer='rmsprop')

early_stop = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=1, mode='auto')

history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=500,
    batch_size=None,
    verbose=2,
    callbacks=[early_stop]
)
```
