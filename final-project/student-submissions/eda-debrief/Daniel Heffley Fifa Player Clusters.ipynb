```python
players = players[players['player_positions'] != 'GK']
```

```python
players.dropna(axis = 1, inplace = True)
```

```python
list(zip(players.columns,players.dtypes))
```

```python
player_names = players['long_name']#Save the names; we need these for later; we will drop all other non-numerical columns
```

```python
players = players.select_dtypes(include = 'number')#keep only numerical columns
```

```python
list(zip(players.columns,players.dtypes))
```

```python
players = players.drop(['sofifa_id','age'],axis =1)#The id and age aren't important
```

```python
list(zip(players.columns,players.dtypes))
```

```python
drop_list = []#Get rid of all the goalkeeping stats as the goalkeepers were dropped
for col in players.columns:
    if 'goalkeeping' in col:
        drop_list.append(col)
    
```

```python
drop_list
```

```python
players = players.drop(drop_list,axis =1)
```

```python
list(zip(players.columns,players.dtypes))
```

```python
players['Name'] = player_names
```
