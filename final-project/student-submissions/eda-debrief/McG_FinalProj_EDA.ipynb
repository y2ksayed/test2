```python
#plotting all the Datapoints related to Financial Services industry because I wanted to brake out the relevant groups
#for the most part, metrics seem to move in similar patterns. 
MC[['FS_Favorable_per', 'FS_Unfav_per', 'FS_Neg_per','FS_Pos_per','FS_Distrust_per', 'FS_Trust_per',
    'FS_admiredEmp_per','FS_PosBuzz_per']].plot(figsize=(15,8));
```

```python
#plotting all the Datapoints related to Insurance industry 
MC[['Ins_Favorable_per', 'Ins_Unfav_per', 'Ins_Neg_per','Ins_Pos_per','Ins_Distrust_per', 'Ins_Trust_per',
    'Ins_admiredEmp_per','Ins_PosBuzz_per']].plot(figsize=(15,8));
```

```python
#plotting all the Datapoints related to Prudential 
MC[['PRU_Favorable_per', 'PRU_Unfav_per', 'PRU_Neg_per','PRU_Pos_per','PRU_Distrust_per', 'PRU_Trust_per',
    'PRU_admiredEmp_per','PRU_PosBuzz_per']].plot(figsize=(15,8));
```

```python
#plotting all the additional Datapoints related to Prudential
MC[['Stock']].plot(figsize=(15,5)); #stockprice has fell since Dec 2017 with two dramatic dips 
MC[['Twitter_Mentions', 'Media_Mentions']].plot(figsize=(15,5));#news mentions were relatively flat until Oct 2018; Twitter is mostly lower than news except for a few spikes 
MC[['Twitter_Sentiment', 'Media_Sentiment']].plot(figsize=(15,5)); #sentiment doesn't fluctuate as much in 2019 as it did in 2018 
```

```python
#When comparing the Trust ranking across Pru and the two industry data, 
#Pru is mostly higher than both with Financial services the lowest. 

plot_1 = MC[['PRU_Trust_per', 'FS_Trust_per', 'Ins_Trust_per']].plot(figsize=(15,8));
plot_1.set_title('Trust', fontsize=30, y=1.01);
plot_1.legend(loc=3);
```

```python
#distrust in prudential is higher than in the industry metrics as well 
plot_1 = MC[['PRU_Distrust_per', 'FS_Distrust_per', 'Ins_Distrust_per']].plot(figsize=(15,8));
plot_1.set_title('Distrust', fontsize=30, y=1.01);
plot_1.legend(loc=3);
```

```python
plot_1 = MC[['PRU_Favorable_per', 'FS_Favorable_per', 'Ins_Favorable_per']].plot(figsize=(15,8));
plot_1.set_title('Favorability', fontsize=30, y=1.01);
plot_1.legend(loc=3);
```

```python
plot_1 = MC[['PRU_admiredEmp_per', 'FS_admiredEmp_per', 'Ins_admiredEmp_per']].plot(figsize=(15,8));
plot_1.set_title('Admired Employer', fontsize=30, y=1.01);
plot_1.legend(loc=3);
```

```python
#Mapping out all the Prudential NEGATIVE attributes 
#There doesn't seem to be any connection between Prudential's stock and any of the other attributes. 
#there might be something between stock price and twitter sentiment 

pd.plotting.scatter_matrix(MC[['PRU_Unfav_per', 'PRU_Neg_per', 'PRU_Distrust_per', 'Stock', 'Twitter_Sentiment', 
                               'Media_Sentiment', 'Twitter_Mentions', 'Media_Mentions']], figsize=(20, 15));
```
